```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import eigh
```

```python
e, hbar, m = 3.795, 1973, 0.511e6
h2m = hbar**2 / (2 * m)

N, r_max = 1000, 20
dr = r_max/N
r = np.linspace(dr, r_max, N)
```

```python
a = 7
V = (-e**2 / r) * np.exp(-r/a)
T = np.array([np.array([-2 if i==j else 1 if abs(i-j) == 1 else 0 for j in range(N)]) for i in range(N)])
H = - h2m * T / dr**2 + np.diag(V)
```

```python
val, vec = eigh(H)
n = 4

# print 1st n values
print(val[:n])

plt.figure(figsize=(8, 5))
for i in range(n):
    E = val[i]
    psi = vec[:, i] / np.sqrt(dr)
    plt.plot(r, psi**2, label = f"E = {E:.3f} eV")

plt.xlabel("r")
plt.ylabel("$|\psi(r)|^2$")
plt.legend()
plt.title(f"Radial Probability Density of Hydrogen Atom for a = {a}")
plt.grid()
plt.show()
```

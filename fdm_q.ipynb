```python
import numpy as np
import matplotlib.pyplot as plt

start, end, N = -6, 6, 500
x = np.linspace(start, end, N)
dx = x[1]-x[0]
```

```python
hbar, mass = 1, 1
# particle in a box
pot = lambda x: 0
# particle in a finite potential well
pot = lambda x: 0 if x < 3 and x > -3 else 5
# particle in harmonc potential
pot = lambda x: x**2

plt.plot(x, [pot(i) for i in x])
```

```python
T = np.array([np.array([-2 if i==j else 1 if abs(i-j) == 1 else 0 for j in range(N-2)]) for i in range(N-2)])
V = np.array([np.array([pot(x[i+1]) if i==j else 0 for j in range(N-2)]) for i in range(N-2)])

H = -0.5 * T/dx**2 + V
```

```python
val, vec=np.linalg.eig(H)
sorted_index = np.argsort(val)
val /= val[sorted_index][0]

# plotting (scaled down) potential
plt.plot(x, [0.005* pot(i) for i in x])
# plotting wavefunctions
for i in range(5):
    plt.plot(x[1:-1], vec[:,sorted_index[i]], lw=1, label=f'n = {i+1}')
plt.legend()
plt.ylim([-.2,.2])
plt.show()
```
